```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
#Countplot : Survival based on gender
sns.countplot(df, x ='survived',hue='sex', )#palette ='Rbdu (palette is used for coloring)
plt.title('Survived vs nNot Survived by Gender')
plt.show()
```

```python
sns.barplot(df, x = 'class',y='fare', hue='sex')
plt.title('Average Fare by Passenger Class and Gender')
plt.show()
```

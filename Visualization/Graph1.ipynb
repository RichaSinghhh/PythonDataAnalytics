```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
#Countplot : Survival based on gender
sns.countplot(df, x ='survived',hue='sex', )#palette ='Rbdu (palette is used for coloring)
plt.title('Survived vs nNot Survived by Gender')
plt.show()
```

```python
sns.barplot(df, x = 'class',y='fare', hue='sex')
plt.title('Average Fare by Passenger Class and Gender')
plt.show()
```

```python
df.columns
```

```python
sns.catplot(df, x="sex", y="age",
kind="violin", split=True,
)
plt.show()
```

```python
sns.violinplot(df, x="class", y="age", hue="sex")
```

```python
sns.pointplot(df, x='class',y='survived', hue='sex',palette='dark', markers=['o','x'])
plt.title('Survival Rate by Class and Gender')
plt.show()
```
